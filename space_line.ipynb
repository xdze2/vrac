```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# force field
def g(epsilon):
    return G*M/D**2*( epsilon**(-2) - mu/(1 - epsilon)**2 + mu - (1+mu)*epsilon )

def eta(epsilon):
    return 1/epsilon + mu/(1 - epsilon) - mu*epsilon + (1+mu)*epsilon**2/2
```

```python
M = 5.972e24 # kg, mass of Earth
m = 7.348e22 # kg, mass of Moon
D = 384400e3 # m, Earth-Moon distance
R = 6371e3 # m, Radius of the earth
r = 1737e3 # m, Radius of the Moon
mu = m/M

G = 6.64730e-11 # m3kg-1s-2, Gravitational constant
```

```python
eps_span = np.linspace(R/D, 1-r/D, 73)
plt.semilogy(eps_span, np.abs(g(eps_span)));
plt.xlabel('epsilon: Earth --> Moon');
plt.ylabel('g(epsilon)');
plt.axvline(x=R/D); plt.axvline(x=1-r/D);
plt.axvline(x=.12, linestyle=':');
plt.axvline(x=.85, linestyle=':');
```
